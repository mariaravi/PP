# AI Worker Burnout Dataset Cleaning

## Data Cleaning and Type Optimization



# 1. Import Libraries and Load Raw Data



```python
# Import pandas for data cleaning and preprocessing

import pandas as pd

# Load the raw dataset

df = pd.read_csv("../data/raw/ai_worker_burnout_attrition_2026.csv")

# Display the first rows

df.head()
```

# 2. Initial Dataset Check



```python
# Display dataset dimensions

df.shape
```

```python
# Review column names, data types, and non-null values

df.info()
```

```python
# Check missing values before cleaning

df.isna().sum()
```

```python
# Check duplicate rows before cleaning

df.duplicated().sum()
```

# 3. Memory Usage Before Optimization



```python
# Calculate memory usage before type optimization

memory_before = df.memory_usage(deep=True).sum() / 1024

print(f"Memory before: {memory_before:.2f} KB")
```

# 4. Convert Object Columns to Category

Text-based columns with repeated values are converted from `object` to `category`.



```python
# Convert object columns to category dtype

object_cols = df.select_dtypes(include="object").columns

for col in object_cols:
    df[col] = df[col].astype("category")

object_cols
```

# 5. Downcast Integer Columns

Integer columns are downcast to smaller integer types where possible.

This reduces memory usage while keeping the same numerical information.

```python
# Downcast integer columns to reduce memory usage

int_cols = df.select_dtypes(include="int64").columns

for col in int_cols:
    df[col] = pd.to_numeric(df[col], downcast="integer")

int_cols
```

# 6. Downcast Float Columns

Float columns are downcast to smaller float types where possible.

This helps reduce memory usage while preserving the values needed for analysis.

```python
# Downcast float columns to reduce memory usage

float_cols = df.select_dtypes(include="float64").columns

for col in float_cols:
    df[col] = pd.to_numeric(df[col], downcast="float")

float_cols
```

# 7. Memory Usage After Optimization

The memory usage is calculated again after type conversion.

This shows how much memory was saved through data type optimization.

```python
# Calculate memory usage after type optimization

memory_after = df.memory_usage(deep=True).sum() / 1024

print(f"Memory after: {memory_after:.2f} KB")
print(f"Reduction: {(1 - memory_after / memory_before) * 100:.2f}%")
```
