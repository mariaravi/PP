# Respondent ID Consistency

Before merging tables check whether `respondent_id` is present and consistent across all raw files.



```python
# Compare the number of unique respondent IDs across all tables

print("respondents:", respondents["respondent_id"].nunique())
print("usage:", usage["respondent_id"].nunique())
print("salary:", salary["respondent_id"].nunique())
print("skills:", skills["respondent_id"].nunique())
print("ethics:", ethics["respondent_id"].nunique())
```

```python
# Check table size and respondent ID uniqueness for one-row-per-respondent tables

for name, table in {
    "respondents": respondents,
    "salary": salary,
    "skills": skills,
    "ethics": ethics
}.items():
    print(name, table.shape, table["respondent_id"].nunique())
```

# Review Summary

The initial dataset review showed that the raw data is organized across several related tables.

Main observations:

- `respondent_id` is the key variable for connecting the tables.
- Most tables appear to contain respondent-level information.
- The AI tool usage table requires special attention because it may contain multiple records per respondent.
- Missing values and categorical variables should be reviewed carefully during cleaning.
